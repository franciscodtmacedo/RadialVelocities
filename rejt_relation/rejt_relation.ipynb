# How does the rejt value afects the RV

In this document we calculate the RV for diferent rejt values (from 0.990 to 0.999 with a 0.001 interval between steps). The rejt parameter is an input parameter in ARES related to the continuum calibration. For simplicity, we used the spectre of the Sun wich is included in ARES.

```python
import numpy as np
from numpy import loadtxt
import scipy.constants as cnt
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20, 10)

lines = loadtxt('0.ares' , float) # loads the file with the line data
ori_lambda = lines[:,0] # array with the reference values of the wavelength

rejt = np.array([0.990,0.991,0.992,0.993,0.994,0.995,0.996,0.997,0.998,0.999]) # array with the numbers used to access the files

N = len(rejt) # number of diferent rejt values
M = len(ori_lambda) # number of lines

data = np.zeros([N,M]) # empty matrix to store the fitted values of the wavelenghts
# the previous matrix has 27 columns because it is the maximum value of wavelenghts found
RV_matrix = np.zeros([N,M]) # empty matrix to store the RV values for each wavelenght
RV = np.zeros(N) # empty array to store the RV values
uncert_matrix = np.zeros([N,M]) # empty matrix to store the error for each wavelenght
error = np.zeros(N) # empty matrix to store the error for each wavelenght

for i,n in enumerate(rejt):
    
    info=loadtxt('{}.ares'.format(i),float) # loads the file with the data
    fit=info[:,8]
    ref=info[:,0]
    err=info[:,9]
        
    data[i,:]=fit #stores the data for each document in the data matrix 
    
    RV_matrix[i,:] = (fit - ref) / ref * cnt.c # calculates the values of RV and stores them in the RV_matrix
    RV[i] = np.mean(RV_matrix[i,:]) # calculates the mean RV for each rejt value
    uncert_matrix[i,:] = err / ref *cnt.c # calculates the error for each wavelenght and stores it in the uncert_matrix
    error[i] = np.std(RV_matrix[i,:]) / np.sqrt(M) # calculates the standard deviation of RV for each rejt value
    
    plt.plot(ref,RV_matrix[i,:M],'.') # plots the graphic
    
plt.ylabel('Radial Velocity (m/s)')
plt.show()
```

In the above graphic we can see RV for each of the selected wavelenghts. Each colour represents a diferent rejt value, however this representation is not very relevant because it would be hard to make any conclusion out of it. In this regard, in the below graphic we can see the mean RV for each rejt value. From 0.990 to 0.996 it does not appear to change much, but for rejt>0.996 the mean value increases significantly. A possible explanation for this is that the rejt gives the program information of where to consider the continuum to be and, for this spectre, although idealy it should be 1, the best value apears to be between 0.990 and 0.996. Actually, the standard deviation is larger fot rejt>0.996 than fot rejt<0.996 and the smallest value is given for rejt=0.994. If we run ARES and make it calculate the rejt using 3 wavelenght intervals it gives us rejt=0.997048.

```python
plt.plot(rejt,RV,'o')
plt.ylabel('Mean Radial Velocity (m/s)')
plt.xlabel('rejt')
plt.show()

plt.errorbar(rejt,RV,yerr=error,ecolor='black',fmt='o')
plt.ylabel('Mean Radial Velocity (m/s)')
plt.xlabel('rejt')
plt.show()
```

The above graphic looked odd, even more so taking in account the rejt calculated by ARES. Because the values above could be changed by some outliers, the mean RV and error was calculated again removing outliers (criterion for the outliers: be outside of the interval $[\mu - \sigma , \mu + \sigma]$). Now, in the bellow graphic, the results seem more coherent. As a matter of fact, choosing a rejt too close to 1 seems to be a bad decision because if the spectre as considerable noise, fewer values will be used to make the computation, increasing the error.

```python
corrRV = np.zeros(N)
corr_error = np.zeros(N)

for n in range(N):
    newRV = np.zeros(0)
    for i,rv in enumerate(RV_matrix[n,:]): # this cycle chooses the values between the mean and standard deviation
        if RV[n]-error[n]*np.sqrt(M)<rv<RV[n]+error[n]*np.sqrt(M):
            newRV=np.append(newRV,rv)
    corrRV[n]=np.mean(newRV)
    corr_error[n]=np.std(newRV) / np.sqrt(len(newRV))

plt.plot(rejt,corrRV,'o')
plt.ylabel('Mean Radial Velocity (m/s)')
plt.xlabel('rejt')
plt.show()

plt.errorbar(rejt,corrRV,yerr=corr_error,ecolor='black',fmt='o')
plt.ylabel('Mean Radial Velocity (m/s)')
plt.xlabel('rejt')
plt.show()
```
