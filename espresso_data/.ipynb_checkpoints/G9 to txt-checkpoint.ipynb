```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,10)

data = fits.getdata("ESPRESSO_G9.fits")
wave = data["lambda"]
contrast = data["contrast"]

plt.plot(wave,contrast)
```

```python
N=len(wave)
matrix=np.zeros([N,2])
matrix[:,0]=wave
matrix[:,1]=contrast

lines=np.zeros(0)

for i in range(N):
    if matrix[i,0]>4500 and matrix[i,0]<7000:
        if matrix[i,1]>0.20:
            lines=np.append(lines,matrix[i,0])
        else:
            pass
    else:
        pass

lines=np.append(0,lines)
lines=np.append(0,lines)
```

```python
np.savetxt('lines_1030.txt',lines,newline='\n')
```
