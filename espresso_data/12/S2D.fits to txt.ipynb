```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,10)

wave = fits.getdata("r.ESPRE.2021-11-04T02_11_59.535_S2D_A.fits",extname='WAVEDATA_AIR_BARY')
flux = fits.getdata("r.ESPRE.2021-11-04T02_11_59.535_S2D_A.fits",extname='SCIDATA')
```

```python
plt.plot(wave[100],flux[100])
```

```python
M=len(wave)
N=len(wave[1])
for i in range(M):
    matrix=np.zeros([N,2])
    matrix[:,0]=wave[i]
    matrix[:,1]=flux[i]
    np.savetxt('wavelenght_flux_12/{}.txt'.format(i+1,wave[i,0],wave[i,-1]),matrix,newline='\n')
```
