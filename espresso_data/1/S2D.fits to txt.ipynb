```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,10)

wave = fits.getdata("r.ESPRE.2022-07-06T09_27_30.199_S2D_A.fits",extname='WAVEDATA_AIR_BARY')
flux = fits.getdata("r.ESPRE.2022-07-06T09_27_30.199_S2D_A.fits",extname='SCIDATA')
```

```python
for i in range(170):
    plt.plot(wave[i],flux[i])
plt.xlabel('wavelength (A)')
plt.ylabel('flux')
plt.show()
```

```python
for i in range(170):
    plt.plot(wave[i],flux[i])
plt.xlim(5600,5800)
plt.xlabel('wavelength (A)')
plt.ylabel('flux')
plt.show()
```
