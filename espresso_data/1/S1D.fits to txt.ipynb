```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,10)

data = fits.getdata("r.ESPRE.2022-07-06T09_27_30.199_S1D_A.fits")
wave = data["wavelength_air"]
flux = data["flux"]

plt.plot(wave,flux)
plt.xlabel('wavelength (A)')
plt.ylabel('flux')
plt.show()
```
