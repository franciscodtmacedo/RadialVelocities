```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10 , 5)

time = np.array([59766.39409952,59574.01154767,59606.03290452,59623.02443698,59744.37895858,59766.39496427])
order = np.array([5,1,2,3,4,6])
CCF = np.array([-16.6486326637339,-16.646123709192,-16.6472834284326,-16.6466334946945,-16.6482856364796,-16.6496691226057])
std_CCF = np.std(CCF)
error_CCF = np.array([8.3746237634394e-05,0.000108262586539199,9.83589859648827e-05,0.000111158607268523,0.000123709214790978,8.46082588302468e-05])

dispersion = np.zeros([4,3])
```

## 229

```python
RV_mean = np.array([-16651.15397,-16644.44673,-16637.36724,-16623.45697,-16635.81752,-16648.01302])*0.001
RV_wmean = np.array([-16641.76529,-16652.72210,-16623.78036,-16636.73502,-16635.90425,-16651.45129])*0.001
RV_median = np.array([-16643.34339,-16642.79447,-16635.92343,-16635.54253,-16626.39689,-16645.14565])*0.001
error_RV_mean = np.array([12.76368,20.18096,18.91836,15.26146,16.27901,14.54402])*0.001
error_RV_wmean = np.array([2.02293,1.51089,1.50283,1.52154,1.60981,1.52261])*0.001

dispersion[0,0] = np.std(RV_mean)
dispersion[0,1] = np.std(RV_wmean)
dispersion[0,2] = np.std(RV_median)
```

```python
plt.errorbar(order,CCF-RV_mean,yerr=error_RV_mean,ecolor='blue',fmt='o',label='mean')
plt.errorbar(order,CCF-RV_wmean,yerr=error_RV_wmean,ecolor='orange',fmt='o',label='wmean')
plt.plot(order,CCF-RV_median,'o',label='median')
plt.title('RV(CCF)-RV , 229 wavelenghts')
plt.xlabel('order')
plt.ylabel('diference / km/s')
plt.legend()
plt.show()

plt.errorbar(time,CCF-RV_mean,yerr=error_RV_mean,ecolor='blue',fmt='o',label='mean')
plt.errorbar(time,CCF-RV_wmean,yerr=error_RV_wmean,ecolor='orange',fmt='o',label='wmean')
plt.plot(time,CCF-RV_median,'o',label='median')
plt.title('RV(CCF)-RV , 229 wavelenghts')
plt.xlabel('time')
plt.ylabel('diference / km/s')
plt.legend()
plt.show()
```
