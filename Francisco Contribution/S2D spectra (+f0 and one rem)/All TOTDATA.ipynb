Having done that the strategy is as follows: Obtain all the rows of all the TotData matrices into a single object and organize them by increasing order of lambda_ref (since each row corresponds to one and only one lambdaref). When this is done, we shall do a cycle that will run through this object and organize it.

```python
AllData = [globals()[f'TotData{i}'] for i in range(1, 21)] #Reunites all the TotData into an array
```

```python
all_rows = [] #Creates a list of rows
for data in AllData: #Going through TotData_i
    for row in data: #Looking through the rows of TotData_i
        key = row[0] #The "key" (used to identify the row) is the first element of the row
        col1, col2, col3, col4 = row[1], row[2], row[3], row[4] 
        all_rows.append((key, col1, col2, col3, col4)) #Adds a tuple that contains information of the row to the python list all_rows
```

We shall now do the following - For a line, we shall subtract the RV value of that line and its correspondant CCF RV for each of the observations, do the average of those diferences and then save it in the 2nd column of a matrix called Results. We shall also calculate the standard deviation of those differences and save it in the third column of that matrix (the first column will have the corresponding reference wavelength)

```python
Results = np.zeros((3417,3))
for i in range(3417) :
    Results[i,0] = TOTDATA[i,0]
    array = (-16600+TOTDATA[i,3]) - TOTDATA[i,4]
    Results[i,1] = np.mean(array)
    Results[i,2] = np.std(array)
Results
```

Having the average and standard deviation of $RV-CCF$ for every line, we can present them through a graph and see what kind of distribution they present. By looking at the distribution, we hope to develop a criterion that can tell us which lines are better and which lines are worse.
